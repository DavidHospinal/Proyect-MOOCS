# Generar números aleatorios de 4 dígitos y luego actualizar los valores de la columna "CantidadVentas" en el DataFrame df4

# este código, utilizamos np.random.randint para generar números enteros aleatorios de 4 dígitos entre 1000 y 9999. Luego, utilizamos el método loc de Pandas para seleccionar las filas donde el valor de "CursoPaga" es False y actualizamos los valores de "CantidadVentas" con los números aleatorios generados.

```python
import numpy as np

# Obtener los índices de las filas donde CursoPaga es False
indices = df4.loc[df4['CursoPaga'] == False].index

# Generar valores aleatorios de 4 dígitos para las filas seleccionadas
random_values = np.random.randint(10,999, size=len(indices))

# Actualizar los valores de CantidadVentas usando los índices
df4.loc[indices, 'CantidadVentas'] = random_values
```

```python
import pandas as pd
import numpy as np

# Filtrar las filas con 'Introductory' en la columna 'Nivel'
df_filtered = df1[df1['Nivel'] == 'Introductory']

# Calcular la cantidad de filas correspondientes al 30%
num_rows = int(len(df_filtered) * 0.3)

# Seleccionar aleatoriamente las filas a las que se les asignará 'False' en 'CursoPaga'
random_rows = np.random.choice(df_filtered.index, size=num_rows, replace=False)

# Actualizar el valor de 'CursoPaga' a 'False' en las filas seleccionadas
df1.loc[random_rows, 'CursoPaga'] = False
```
